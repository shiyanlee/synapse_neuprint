```python
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.layouts import gridplot
from bokeh.models import BoxAnnotation
from bokeh.io import output_notebook
from bokeh.io import output_file

# Display plots inline (if using Jupyter Notebook)
output_notebook()

# Assuming your synapse data is loaded into a DataFrame named `df`
# Replace 'x_pre' and 'z_pre' with the actual column names if different
x_values = df['x_pre']
z_values = df['z_pre']
y_values = df['y_pre']

# Adjust these values as needed
x_min_z, x_max_z = 12309, 17000
z_min_z, z_max_z = 4408, 22796

x_min_y, x_max_y = 12309, 17000
y_min_y, y_max_y = 30000, 37000

z_min_y, z_max_y = 4408, 22796
y_min_y, y_max_y = 30000, 37000

# Create individual plots
p1 = figure(title="x vs z", x_axis_label='x_pre', y_axis_label='z_pre', width=400, height=400)
p1.scatter(x_values, z_values, size=10, color="navy", alpha=0.5)
p1.xaxis.axis_label = 'x_pre'
p1.yaxis.axis_label = 'z_pre'
# Add bounding box to the first plot
box1 = BoxAnnotation(left=x_min_z, right=x_max_z, bottom=z_min_z, top=z_max_z, fill_alpha=0.2, fill_color='red')
p1.add_layout(box1)

p2 = figure(title="x vs y", x_axis_label='x_pre', y_axis_label='y_pre', width=400, height=400)
p2.scatter(x_values, y_values, size=10, color="green", alpha=0.5)
p2.xaxis.axis_label = 'x_pre'
p2.yaxis.axis_label = 'y_pre'
# Add bounding box to the second plot
box2 = BoxAnnotation(left=x_min_y, right=x_max_y, bottom=y_min_y, top=y_max_y, fill_alpha=0.2, fill_color='blue')
p2.add_layout(box2)

p3 = figure(title="z vs y", x_axis_label='z_pre', y_axis_label='y_pre', width=400, height=400)
p3.scatter(z_values, y_values, size=10, color="orange", alpha=0.5)
p3.xaxis.axis_label = 'z_pre'
p3.yaxis.axis_label = 'y_pre'
# Add bounding box to the third plot
box3 = BoxAnnotation(left=z_min_y, right=z_max_y, bottom=y_min_y, top=y_max_y, fill_alpha=0.2, fill_color='green')
p3.add_layout(box3)

# Combine plots into a grid layout
grid = gridplot([[p1, p2], [p3, None]], toolbar_location=None, sizing_mode='fixed')

output_file("my_plots.html")
show(grid)

# # Calculate the max and min values for pre-synaptic coordinates
# max_x_pre = x_values.max()
# min_x_pre = x_values.min()
# max_z_pre = z_values.max()
# min_z_pre = z_values.min()
# min_y_pre = y_values.min()
# max_y_pre = y_values.max()

# # Define the bounding box range 
# min_x_box, max_x_box = min_x_pre, max_x_pre
# min_z_box, max_z_box = min_z_pre, max_z_pre
# max_y_box, max_y_box = mine

# # Create a Bokeh plot with customized size
# p = figure(title="Synapse Locations with Bounding Boxes", 
#            x_axis_label='x_pre', y_axis_label='z_pre',
#            width=800,  # Set the width of the plot
#            height=600)  # Set the height of the plot

# # Plot the pre-synaptic coordinates (x_pre vs z_pre)
# p.scatter(x_values, z_values, size=10, color="navy", alpha=0.5)

# # Add the first bounding box 
# box1 = BoxAnnotation(left=min_x_box, right=max_x_box, 
#                     bottom=15000, top=max_z_box, 
#                     fill_alpha=0.2, fill_color='red')
# p.add_layout(box1)

# # Add the second bounding box 
# box2 = BoxAnnotation(left=14000, right=17285, 
#                     bottom=min_z_box, top=16000, 
#                     fill_alpha=0.2, fill_color='green')
# p.add_layout(box2)

# # Customize the x-axis ticks
# p.xaxis.ticker = [i for i in range(int(min_x_pre), int(max_x_pre) + 1000, 1000)]  # Example: major ticks every 1000 units
# p.xaxis.major_label_orientation = "vertical"  # Rotate labels if needed

# p.xgrid.grid_line_color = "gray"
# p.xgrid.grid_line_dash = "dotted"

# # Optionally flip the y-axis if needed
# p.y_range.flipped = True

# # Show the plot
# show(p)
```
